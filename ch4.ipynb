### Python based method'

### Linear reg applies to numerical values

### Step 1
- Import libraries

````python
import numpy as np
from sklearn.linear_model import LinearRegression
````

### Step 2: - Import dataset

````python
x=np.array([60,62,67,70,71,72,75,78]).reshape(-1,1)
print(x)
````

````python
y=np.array([140,155,159,179,192,200,212,215])
print(y)
````

### Step 3: Create a model

````python
model = LinearRegression()
````

### Step 4: Fit the model


````python
model.fit(x,y)
````

### Step 5: Print the slope and intercept

````python
#slope
print(model.coef_[0])
#intercept
print(model.intercept_)
````

````python
slope = model.coef_[0]
intercept = model.intercept_
````

````python
print(f'The linear regression equation is y = {intercept:.2f}+{slope:.2f}*x')
````

### Prediction

````python
model.predict([[90]])
````

### Polynomial regression
- it is used when the data is not linear but follows curved patttern



````python
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
````

````python
# load dataset
df = pd.read_csv("Iris(1).csv")
````

````python
x = df[['SepalLengthCm']]
y = df['PetalLengthCm']
print(x.shape)
print(y.shape)
````

### create polynomial features

````python
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)
model = LinearRegression()
model.fit(x_poly,y)
print(model.coef_) 
print(model.intercept_)
````
