````python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
````

### Prepare dataframe

````python
iris = load_iris()
df = pd.DataFrame(iris.data,columns=iris.feature_names)
print(df)
df['target'] = iris.target
df['species'] = df['target'].map({
    0:"setosa",
    1:"versicolor",
    2:"virginica"
})
df.head(2)
````

### `pd.get_dummies`
- this method is used to convert categorical variables to boolean equivalents
- this conversion facilitates ml ana

````python
pd.get_dummies(df,columns=["species"],drop_first=True)
````

### `np.where`
- this is used to apply conditions to the given columns of df

````python
df["size"]=np.where(df["sepal length (cm)"]>6.5,"Large",
           np.where(df["sepal length (cm)"]>5.5,"Medium",
           "Small"))
````
