### CLassification models vs regression models


### Classification
- Data: Divided into predefined, discrete categories
- Goal: Assign an input to one of the categories
- Model output: A specific category from a set

### Regression
- Data: Continuous numerical data with an underlying relationship
- Goal: Predict one variable based on others
- Model output: A numerical value

### Decision tree
- a tree like rule based system
- used for classification as well as regression

### Python implementation of job offers using decision tree

````python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier,plot_tree
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df = pd.read_csv("job_offers.csv")
df.head(2)                                                                                                                                                                                  
````

````python
### Define features and targets
x = df.drop(columns=['JobOffered'])
y = df['JobOffered']
print(x.shape)
print(y.shape)
````

````python
# create model
dtree = DecisionTreeClassifier(criterion='entropy',max_depth=3)
````

````python
# fit the model
dtree.fit(x,y)
````

````python
# prediction
y_pred = dtree.predict(x)
print(y_pred)
````

````python
# plot the tree (Not in  Syllabus)
plt.figure(figsize=(12,14))
plot_tree(dtree,filled=True,feature_names=['CGPA','Communiaction','Aptitude','ProgrammingSkill'],class_names=['No','Yes'])
plt.title("Decision tree of job offered")
plt.show()
````

### Evaluation metrics in classification model
- confusion metrics
- accuracy score
- recall score
- f1 score
- classification report

````python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,classification_report
````

### Confusion Matrix

````python
cm = confusion_matrix(y,y_pred)
print("Confusion matrix \n",cm)
# TN,FP,FN,TP = cm.ravel()
````

````python
TN,FP,FN,TP = cm.ravel()
print("True Negatives:",TN)
print("False Positives:",FP)
print("False Negatives:",FN)
print("True Positives:",TP)
````

# Accuracy score
- accuracy = (TN+TP) / (TN+TP+FN+FP)

````python
accuracy = (TN+TP) / (TN+TP+FN+FP)
print(accuracy)
# alternate
print(accuracy_score(y,y_pred))
````

# PrecisionnnnnNNNNNNNNNNNNNNNN
- precision = TP / (TP+FP)

````python
precision = TP / (TP+FP)
print(precision)
# alternate
print(precision_score(y,y_pred))
````

### RECALL
- recall = TP / (TP+FN)

````python
recall = TP / (TP+FN)
print(recall)
#alternate
print(recall_score(y,y_pred))
````

### F1score
- f1score = 2 * (precision*recall) / (precision+recall)

````python
f1score = 2 * (precision*recall) / (precision+recall)
print(f1score)
# alternate
print(f1_score(y,y_pred))
````

#  Classification report

````python
print(classification_report(y,y_pred))
````

### Plotting confusion matrix heatmap

````python
plt.figure(figsize=(6,5))
sns.heatmap(cm,
           annot=True,
           fmt='d',
           cmap='Blues',
           xticklabels=['No','Yes'],
           yticklabels=['No','Yes'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Heatmap of confusion matrix")
plt.show()
````
